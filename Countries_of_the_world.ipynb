## <b>Analysis of 227 Countries Dataset Consists of Various Properties Like Population, Area, Region, etc.</b>
#### The core focus of this analysis is to identify most densed countries in the world, how population is distributed among different regions, which are the countries where population has increased and decreased most in a year.

```python
import pandas as pd
import numpy as np
df=pd.read_csv("countries of the world.csv")
df.head()
```

#### Let's look at the datatypes of columns present in dataset

```python
df.dtypes
```

#### As we can see in the dataset, some of the numeric values are comma separated and they are as string.
#### So, in order to perform analytical operation on them, we need to convert comma separated values to proper floating type values and convert datatype of them.

```python
column_names=np.array(['Pop. Density (per sq. mi.)','Coastline (coast/area ratio)','Net migration','Infant mortality (per 1000 births)','Literacy (%)','Phones (per 1000)','Arable (%)','Crops (%)','Other (%)','Birthrate','Deathrate','Agriculture','Industry','Service'])
for column in column_names:
    df[column]=df[column].str.replace(',','.')
    df[column]=df[column].astype("float")
df.head()
```

#### As we can see below, datatypes of certain columns has been changed as we expected

```python
df.dtypes
```

### <b>For performing any analysis on data, data has to be clean. For cleaning this dataset, null or empty values has to be handled in such a way that it should not affect our analysis.</b>
#### Let's look at all the null values present in our dataset by columns

```python
missing_data = df.isnull()

for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")
```

#### Here we are replacing <i>NaN</i> values with 0 in numeric columns for getting rid of <i>NaN</i> values from dataset

```python
null_columns=np.array(['Net migration','Infant mortality (per 1000 births)','GDP ($ per capita)','Literacy (%)','Phones (per 1000)','Arable (%)','Crops (%)','Other (%)','Climate','Birthrate','Deathrate','Agriculture','Industry','Service'])
for column in null_columns:
    df[column].replace(np.nan,0,inplace=True)
    
df.head()
```

#### Certain fields has to be in complete number form without decimal point, so will change some fields to int type with rounded value

```python
valuesToRoundOff=np.array(['Pop. Density (per sq. mi.)','Coastline (coast/area ratio)','Infant mortality (per 1000 births)','GDP ($ per capita)','Literacy (%)','Phones (per 1000)','Birthrate','Deathrate'])
df[valuesToRoundOff]=df[valuesToRoundOff].apply(lambda value: round(value))
for column in valuesToRoundOff:
    df[column]=df[column].astype("int")
df.head()
```

#### Getting Top 10 most densed countries

```python
sorted_by_density = df[['Country','Area (sq. mi.)','Pop. Density (per sq. mi.)']].sort_values(by='Pop. Density (per sq. mi.)', ascending=False)
top10_densed_countries=sorted_by_density.head(10)
top10_densed_countries
```

#### As we've seen in the result, <b>Monaco</b> is the most densed country in the world with 16272 people living in per square meter area, followed by <b>Macau</b> and <b>Singapore</b> to make Top 3 most densed countries in the world.
#### Plotting data on bar chart

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('bmh')

ax=top10_densed_countries.plot(kind='barh',x='Country',figsize=(20,12),title='Most Densed Countries By Area/Population Density (Per Sqr. mi)', fontsize=15)

for i in ax.patches:
    ax.text(i.get_width()+150, i.get_y()+0.06, \
            str(round((i.get_width()), 2)), fontsize=13, color='blue')
    
top10_densed_countries
```

#### Getting all the region names with number of countries associated with it

```python
df_pop=df[['Country','Population','Region']]
df_pop.groupby(['Region'])['Country'].count()
```

#### Calculating total population of region by sum up population of all countries associated with particular region

```python
total_population=df_pop.groupby(['Region'])['Population'].apply(sum)
print(total_population)

population_by_region=pd.DataFrame()

population_by_region=population_by_region.assign(Region=df_pop['Region'].unique().astype(str))
population_by_region=population_by_region.sort_values('Region')

population_by_region['Total Population']=total_population.values
population_by_region
```

#### As we have entire data of population distribution with us, so now will plot it on Pie chart to see how it is distributed.

```python
region_names=['ASIA (EX. NEAR EAST)','BALTICS','C.W. OF IND. STATES','EASTERN EUROPE','LATIN AMER. & CARIB','NEAR EAST','NORTHERN AFRICA','NORTHERN AMERICA','OCEANIA','SUB-SAHARAN AFRICA','WESTERN EUROPE']
population_by_region.plot.pie('Total Population',figsize=(17,17),fontsize=15,title="POPULATION DISTRIBUTION BY REGIONS",labels=region_names,autopct='%.1f%%',legend=False,shadow=False,pctdistance=0.93,startangle=100)
plt.legend(loc='upper left',fontsize=12)
```

#### In the dataset we have Infant mortality, deathrate, and birthrate. Based on these three values we can get change in population in a year for a country

```python
df['change_in_population']=df['Infant mortality (per 1000 births)']+df['Deathrate']-df['Birthrate']
df.head()
```

#### As we show above, change in population field has been appended in dataset, so now will take top 10 countries with positive change and negative change in population and plot them on bar chart.

```python
ax=df[['Country','change_in_population']].sort_values(by='change_in_population',ascending=False).head(10).plot(kind='barh',figsize=(12,5),fontsize=14,x='Country',title='Top 10 Countries Having Increment in Population Per 1000 People In a Year',legend=False)

for i in ax.patches:
    ax.text(i.get_width()+1, i.get_y()+0.07, \
            str(round((i.get_width()), 2)), fontsize=13, color='blue')
```

```python
ax=df[['Country','change_in_population']].sort_values(by='change_in_population',ascending=False).tail(10).plot(kind='barh',figsize=(12,5),fontsize=14,x='Country',title='Top 10 Countries Having Decrement in Population Per 1000 People In a Year',legend=False)

for i in ax.patches:
    ax.text(i.get_width()-0.7, i.get_y()+0.07, \
            str(round((i.get_width()), 2)), fontsize=13, color='blue')
```
