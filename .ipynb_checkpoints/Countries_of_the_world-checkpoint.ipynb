```python
import pandas as pd
import numpy as np
df=pd.read_csv("countries of the world.csv")
df.head()
```

```python
df.dtypes
```

```python
#change dtype obj to str
column_names=np.array(['Pop. Density (per sq. mi.)','Coastline (coast/area ratio)','Net migration','Infant mortality (per 1000 births)','Literacy (%)','Phones (per 1000)','Arable (%)','Crops (%)','Other (%)','Birthrate','Deathrate','Agriculture','Industry','Service'])
for column in column_names:
    df[column]=df[column].str.replace(',','.')
    df[column]=df[column].astype("float")
df.head()
```

```python
df.dtypes
```

```python
missing_data = df.isnull()
missing_data.head(5)
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")
```

```python
df.head()
```

```python
null_columns=np.array(['Net migration','Infant mortality (per 1000 births)','GDP ($ per capita)','Literacy (%)','Phones (per 1000)','Arable (%)','Crops (%)','Other (%)','Climate','Birthrate','Deathrate','Agriculture','Industry','Service'])
for column in null_columns:
    df[column].replace(np.nan,0,inplace=True)
    
df.head()
```

```python
valuesToRoundOff=np.array(['Pop. Density (per sq. mi.)','Coastline (coast/area ratio)','Infant mortality (per 1000 births)','GDP ($ per capita)','Literacy (%)','Phones (per 1000)','Birthrate','Deathrate'])
df[valuesToRoundOff]=df[valuesToRoundOff].apply(lambda value: round(value))
for column in valuesToRoundOff:
    df[column]=df[column].astype("int")
df.head()
```

```python
df.describe()
```

```python
sorted_by_density = df[['Country','Area (sq. mi.)','Pop. Density (per sq. mi.)']].sort_values(by='Pop. Density (per sq. mi.)', ascending=False)
sorted_by_density.head()
```

```python
def CalculatePeoplePerSqrMtr(area, density):
    if area == 0 or density == 0:
        return 0
    elif area > density:
        return area/density
    else:
        return density/area
    
PeoplePerSqrMtr=pd.Series([])

for index, row in sorted_by_density.iterrows():
    PeoplePerSqrMtr[row['Country']]=round((CalculatePeoplePerSqrMtr(row['Area (sq. mi.)'], row['Pop. Density (per sq. mi.)'])))
    
df1=pd.DataFrame(PeoplePerSqrMtr, columns=['People (per sq. mi.)'])
countryColumns=df1.index

df1.index=range(len(df))
df1['Country']=countryColumns.astype('str')

sorted_by_density=pd.merge(sorted_by_density, df1, left_on='Country', right_on='Country')
sorted_by_density.drop(['Area (sq. mi.)','Pop. Density (per sq. mi.)'],axis=1,inplace=True)
sorted_by_density.head()
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
%matplotlib inline
import matplotlib.pyplot as plt
 
countries=np.array(sorted_by_density.Country)
y_pos = np.arange(len(countries))
people = np.array(sorted_by_density['People (per sq. mi.)'])
 
plt.barh(y_pos, people, align='center', alpha=0.7)
plt.yticks(y_pos, countries)
plt.xlabel('People Liveing Per Square Meter')
plt.title('Countries')
 
plt.show()
```
